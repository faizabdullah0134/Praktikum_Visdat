```python
import matplotlib.pyplot as plt
```

**Dataset**

```python
# Buat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

**Single Line Plot**

```python
# Line plot 1 garis, style grafik default
plt.plot(months, product_A_sales, label='Product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Kostumisasi Line Plot**

```python
# Customization Line Plot
plt.plot(months, product_A_sales, label='Product A', color='blue', linestyle='--', marker='o')
plt.plot(months, product_B_sales, label='Product B', color='red', linestyle='-', marker='x')
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
```

**Menambahkan Multiple Lines**

```python
# Multiple Line Plot
plt.plot(months, product_A_sales, label='Product A', marker='o')
plt.plot(months, product_B_sales, label='Product B', marker='x')
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.show()
```

**Jenis Garis untuk Menunjukkan Trend**

```python
# Data samole tambahan
product_C_sales = [18,22,25,28,32,38,42,45,48,52,56,60]
product_D_sales = [7,9,11,13,16,18,20,23,25,28,30,33]

# Plot Produk A
plt.plot(months, product_A_sales, label='Product A Trend', linestyle='--', color='blue')

# Plot Produk B
plt.plot(months, product_B_sales, label='Product B Trend', linestyle='-', color='red')

# Plot Produk C
plt.plot(months, product_C_sales, label='Product C Trend', linestyle=':', color='green')

# Plot Produk D
plt.plot(months, product_D_sales, label='Product D Trend', linestyle='-.', color='purple')

plt.title('Tren Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplot untuk Menampilkan Beberapa Grafik**

```python
# Line Plot
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot for Product A
axs[0].plot(months, product_A_sales, label='Product A', color='blue', marker='o')
axs[0].set_title('Penjualan Produk A Per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# Plot for Product B
axs[1].plot(months, product_B_sales, label='Product B', color='red', marker='x')
axs[1].set_title('Penjualan Produk B Per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
